# What is the distribution of the fall of wickets wrt the over being bowled?

```python
df['player_dismissed'].unique()
```

```python
 df['wicket_fell'] = df.player_dismissed.notna()
```

```python
df.head()
```

```python
df_Wickets = df[df.wicket_fell]
df_Wickets
```

```python
wickets_by_over_count = df_Wickets['over'].value_counts(sort=True, normalize=True) * 100
wickets_by_over_count 
```

```python
wickets_by_over_count = wickets_by_over_count.sort_index()
wickets_by_over_count
```

```python
wickets_by_over_count.plot.bar(color='blue')
plt.show()
```

# Which teams score the most in the last 4 overs?

```python
df_over_17plus = df[df.over >= 17]
df_over_17plus.head()
```

```python
df_over_17plus_grp = df_over_17plus.loc[:,['match_id','over','batting_team','total_runs']].groupby(['match_id','over','batting_team']).sum()
df_over_17plus_grp.head(20)
```

```python
df_over_17plus_grp.groupby(level=2).mean().sort_values(by='total_runs', ascending=False)
```
